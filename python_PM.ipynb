```python
import investpy
import pandas as pd
```

```python
search_results = investpy.search_etfs(by="name", value="ishares us")
```

```python
# liste pour faciliter la recherche des noms...
search_results[search_results["country"] == "united states"]
```

```python
liste_etf = ["iShares us Financials", "iShares us Telecommunications", "iShares US Industrials", 
             "iShares US Consumer Services", 
             "iShares US Consumer Goods", "iShares US Utilities", "iShares US Basic Materials","iShares US Energy"]

# pour les boucle je verais quand j’aurais plus de temps...

# code dégueu mais rapide
financials = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares us Financials", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))
telecom = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares us Telecommunications", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))
industrials = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares US Industrials", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))
cons_services = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares US Consumer Services", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))
cons_goods = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares US Consumer Goods", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))
utilities = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares US Utilities", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))
materials = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares US Basic Materials", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))
energy = pd.DataFrame(investpy.etfs.get_etf_historical_data("iShares US Energy", "united states", "01/01/2010", "01/01/2021", stock_exchange=None, as_json=False, order='ascending', interval='Daily'))

```

```python
# Bien mettre ce fichier "python_PM.ipynb" dans le dossier souhaité pour que
# le téléchargement des fichiers se fasse sans problème

financials.to_excel("financial.xlsx")
telecom.to_excel("telecom.xlsx")
industrials.to_excel("industrials.xlsx")
cons_services.to_excel("cons_services.xlsx")
cons_goods.to_excel("cons_goods.xlsx")
utilities.to_excel("utilities.xlsx")
materials.to_excel("materials.xlsx")
energy.to_excel("energy.xlsx")
```
